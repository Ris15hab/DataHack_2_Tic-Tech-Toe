```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv("LawYantra Dataset - lawyer_summaries.csv")
```

```python
pattern = r'charges(.*?)USD per hour'

# Read the CSV file into a DataFrame
df = pd.read_csv("LawYantra Dataset - lawyer_summaries.csv")

# Create an empty 'Charges' column
df['Charges'] = ""

# Iterate through the 'Information' column, find the text, and append to 'Charges'
for index, info in enumerate(df['Information']):
    matches = re.findall(pattern, str(info))
    if matches:
        df.at[index, 'Charges'] = matches[0].strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)

df['Charges']=df['Charges'].astype(float)
df['Charges']=df['Charges'].round(2)
df['Charges']=df['Charges']*83
```

```python
pattern = r'has(.*?)years of experience'
df['Experience'] = ""
for index, info in enumerate(df['Information']):
    matches = re.findall(pattern, str(info))
    if matches:
        df.at[index, 'Experience'] = matches[0].replace("in has ","").strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)

df['Experience']=df['Experience'].astype(float)
```

```python
pattern = r'has a Client Feedback of (.*?) out of 5.0'
df['Feedback'] = ""
for index, info in enumerate(df['Information']):
    matches = re.findall(pattern, str(info))
    if matches:
        df.at[index, 'Feedback'] = matches[0].replace('N/A','3.0').strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)

df['Feedback']=df['Feedback'].astype(float)
```

```python
pattern = r'speaks:(.*?). He|She'
df['Languages'] = ""
for index, info in enumerate(df['Information']):
    matches = re.findall(pattern, str(info))
    if matches:
        df.at[index, 'Languages'] = matches[0].replace(" and ",", ").strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)

df['Languages']=df.Languages.apply(lambda x: list(set(x.split(','))))
```

```python
pattern = r'practices at (.*?),'
df['Company Name'] = ""
for index, info in enumerate(df['Information']):
    matches = re.findall(pattern, str(info))
    if matches:
        df.at[index, 'Company Name'] = matches[0].strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)
```

```python
pattern = r'experience in (.*?). He|She'
df['Domain'] = ""
for index, info in enumerate(df['Information']):
    matches = re.findall(pattern, str(info))
    if matches:
        df.at[index, 'Domain'] = matches[0].replace("and ","").strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)

df['Domain']=df.Domain.apply(lambda x: list(set(x.split(','))))
```

```python
pattern = r'Jurisdiction is (.*?). He|She'
df['Jurisdiction'] = ""
for index, info in enumerate(df['Information']):
    matches = re.findall(pattern, str(info))
    if matches:
        df.at[index, 'Jurisdiction'] = matches[0].replace("and ","").strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)
```

```python
df['Demographics'] = ""

# Iterate through the 'Information' column, find the text, and append to 'Demographics'
for index, info in enumerate(df['Information']):
    start_str = 'Client Demographics is '
    end_str = '.'
    start_idx = info.find(start_str)
    end_idx = info.find(end_str, start_idx + len(start_str))

    if start_idx != -1 and end_idx != -1:
        demographics = info[start_idx + len(start_str):end_idx]
        df.at[index, 'Demographics'] = demographics.strip()

# Save the modified DataFrame back to the same CSV file
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)
```

```python
df['Days of Disposal'] = ""

# Iterate through the 'Information' column, find the text, and append to 'Days of Disposal'
for index, info in enumerate(df['Information']):
    start_str = 'takes '
    end_str = ' Avg Days for Disposal.'
    start_idx = info.find(start_str)
    end_idx = info.find(end_str, start_idx + len(start_str))

    if start_idx != -1 and end_idx != -1:
        days_of_disposal = info[start_idx + len(start_str):end_idx]
        df.at[index, 'Days of Disposal'] = days_of_disposal.replace('N/A','90.00').strip()

df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)

df['Days of Disposal']=df['Days of Disposal'].astype(float).round()
```

```python
for index, info in enumerate(df['Information']):
    start_str = 'is based in '
    end_str = '.'
    start_idx = info.find(start_str)
    end_idx = info.find(end_str, start_idx + len(start_str))

    if start_idx != -1 and end_idx != -1:
        location = info[start_idx + len(start_str):end_idx]
        df.at[index, 'Location'] = location.strip()
        
df.to_csv("LawYantra Dataset - lawyer_summaries", index=False)
```

```python
df['Flag'] = df['Feedback'].apply(lambda x: 0 if x == 0.0 else 1)
```

```python
def check_pro_bono(text):
    if 'provides' in text and ' pro bono' in text:
        if text.index('provides') < text.index(' pro bono'):
            return 1
    return 0

# Create the 'Pro Bono' column based on the custom function
df['Pro Bono'] = df['Information'].apply(check_pro_bono)
```
